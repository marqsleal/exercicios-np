**Bora praticar!**
  
1) Utilizando o DataFrame importado (alunos3.csv) calcule a média das provas em uma nova coluna chamada (Media_provas)

```python
alunos3 = pd.read_csv('datasets/alunos3.csv', sep=';', decimal=',')
df_alunos3 = pd.DataFrame(alunos3)

df_alunos3['Media_provas'] = df_alunos3[['Prova_1', 'Prova_2', 'Prova_3', 'Prova_4']].mean(axis=1).round(1)
df_alunos3
```

2) Quem foram os alunos que obtiveram a maior e a menor média

```python
aluno_maior_media = df_alunos3.loc[df_alunos3['Media_provas'].idxmax()]
aluno_menor_media = df_alunos3.loc[df_alunos3['Media_provas'].idxmin()]

df_media = pd.DataFrame([aluno_maior_media, aluno_menor_media])

df_media
```

3) Agora una este dataframe com o `cadastro_alunos.xlsx`

```python
df_cadastro_alunos = pd.read_excel('datasets/cadastro_alunos.xlsx')
df_cadastro_alunos3 = pd.merge(df_alunos3, df_cadastro_alunos, on='RA')
df_cadastro_alunos3
```

4) Qual a média entre as Media_provas dentro do público feminino? e masculino?

```python
df_media_por_sexo = df_cadastro_alunos3.groupby('Sexo')['Media_provas'].mean()
df_media_por_sexo
```

5) Qual a média de idade das pessoas que obtiveram Media_provas maior ou igual a 7?

```python
df_media_filtro = df_cadastro_alunos3[df_cadastro_alunos3['Media_provas'] > 7.0]

df_media_por_idade = df_media_filtro['Idade'].mean()
df_media_por_idade.round(2)
```
